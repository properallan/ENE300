Deseja-se maximizar o volume de uma caixa, descrito pela função 

$$
f(x) = 4x^2 - 1014x^2 + 6237x
$$

restrito à

$$
0 \leq x \leq 105
$$

```python
def f(x):
    return (297-2*x)*(210-2*x)*x
```

# Solução usando Busca Dicotîmica

```python
from ene300 import dichotomous_search
```

```python
f_ = lambda x: -f(x)
```

```python
dichotomous_search(f_, 0.0, 105.0 , 0.00001)
```
